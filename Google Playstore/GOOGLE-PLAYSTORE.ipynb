```python
import pandas as pd
google_data=pd.read_csv("DATASETS/googleplaystore.csv",index_col="App")
```

```python
google=google_data.drop_duplicates()
```

```python
google.sample(n=10)
```

```python
chars_in=["M","+","$",",","Free","Everyone","nan","Varies with device","k"]
col_in=["Installs","Price","Size"]

for app in col_in:
    for char in chars_in:
        google[app]=google[app].astype(str).str.replace(char,"")
    google[app]=pd.to_numeric(google[app])
```
